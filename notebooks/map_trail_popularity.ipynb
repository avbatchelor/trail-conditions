```python
import folium
map_1 = folium.Map(location=[44.1644, -71.4326])

folium.CircleMarker([44.1159, -71.6420],
                    radius=5,
                    popup='Mt Liberty',
                    color='#3186cc',
                    fill='TRUE',
                    fill_color='#3186cc',
                   ).add_to(map_1)

map_1
```
